### 1. Bag of Words

```python
import numpy as np
import pandas as pd
```

```python
# creating dataframe
df = pd.DataFrame({"text":["people watch dswithbappy",
                         "dswithbappy watch dswithbappy",
                         "people write comment",
                          "dswithbappy write comment"],"output":[1,1,0,0]})

```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(df['text'])
```

```python
# to know vocabulary
print(cv.vocabulary_)
```

```python
# vectorized matrix using bag of words
bow.toarray()
```

```python
# printing individual row vector matrix
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
```

```python
# using this bow object for new data
cv.transform(["suren watch dswithbabby"]).toarray()
```

```python
X = bow.toarray()
y = df['output']
```
